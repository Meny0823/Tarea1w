# К Algoritmo Gen茅tico en Python

Este cuaderno implementa un algoritmo gen茅tico sencillo en Python, siguiendo el ejemplo del blog de Ander Fern谩ndez. El objetivo es encontrar una cadena de texto igual a una frase objetivo.

Los algoritmos gen茅ticos son t茅cnicas de optimizaci贸n inspiradas en la evoluci贸n biol贸gica, basadas en procesos como **selecci贸n**, **cruzamiento**, y **mutaci贸n**.


```python
import random

#  Par谩metros globales
TARGET = "Hola, soy un algoritmo gen茅tico"
POPULATION_SIZE = 100
MUTATION_RATE = 0.01
GENERATIONS = 1000
CHARACTER_SET = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ ,."

# Funci贸n para generar un car谩cter aleatorio del conjunto
def random_char():
    return random.choice(CHARACTER_SET)

# Crear un individuo (cadena aleatoria)
def create_individual():
    return ''.join(random_char() for _ in range(len(TARGET)))

# Funci贸n de aptitud: cu谩ntos caracteres coinciden con el TARGET
def fitness(individual):
    return sum(individual[i] == TARGET[i] for i in range(len(TARGET)))

# Mutar un individuo aleatoriamente
def mutate(individual):
    individual_list = list(individual)
    for i in range(len(individual_list)):
        if random.random() < MUTATION_RATE:
            individual_list[i] = random_char()
    return ''.join(individual_list)

#  Cruzamiento entre dos padres
def crossover(parent1, parent2):
    pivot = random.randint(0, len(parent1) - 1)
    return parent1[:pivot] + parent2[pivot:]

#  Selecci贸n por torneo
def select(population):
    tournament = random.sample(population, 5)
    tournament.sort(key=lambda ind: fitness(ind), reverse=True)
    return tournament[0]
```

```python
def genetic_algorithm():
    population = [create_individual() for _ in range(POPULATION_SIZE)]

    for generation in range(GENERATIONS):
        population = sorted(population, key=fitness, reverse=True)

        best = population[0]
        print(f"Generaci贸n {generation} | Mejor individuo: {best} | Fitness: {fitness(best)}")

        if fitness(best) == len(TARGET):
            print("\n 隆Soluci贸n encontrada!")
            break

        new_population = population[:10]  # Elitismo

        while len(new_population) < POPULATION_SIZE:
            parent1 = select(population)
            parent2 = select(population)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)

        population = new_population

    print(f"\n Resultado final: {population[0]}")

# Ejecutar el algoritmo
genetic_algorithm()
```

- La tasa de mutaci贸n, el tama帽o de la poblaci贸n y el mecanismo de selecci贸n influyen fuertemente en el rendimiento.
